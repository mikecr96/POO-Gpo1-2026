```python
# nested-if
# if anidado

numero = "HOLA"

if numero.isnumeric():
    if int(numero) % 2 == 0:
        print("Es par")
    else:
        print("Es impar")
else:
    print("NO ES UN STRING NUMÉRICO")
```
