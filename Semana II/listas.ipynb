# CICLOS

## while

```python
i = 0
while i <= 10:
    print("Hola mundo")
    i += 1 # i = i + 1
```

```python
i = 0
while i <= 1000000:
    if i == 10:
        break
    print(i, "Hola mundo")
    i += 1 # i = i + 1
    
```

## FOR

```python
for j in [1, 2, 3]:
    print("Hola mundo", j)
```

```python
for j in "Universidad Panamericana":
    print("Hola mundo", j)
```
