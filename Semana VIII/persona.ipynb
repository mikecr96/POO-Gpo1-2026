## Setters/Getters

```python
class Persona:
    def __init__(self, nombre:str, edad:int) -> None:
        self.__nombre = nombre
        self.__edad = edad
    # Este es getter, sirve para devolver el valor
    def getNombre(self):
        return self.__nombre
    # Este es setter, sirve para modificar el valor
    def setNombre(self, nombre:str):
        self.__nombre = nombre

    # Este es getter, sirve para devolver el valor
    def getEdad(self):
        return self.__edad
    # Este es setter, sirve para modificar el valor
    def setEdad(self, edad:int):
        self.__edad = edad

    def _morir(self):
        print(f"QEPD {self.__nombre}!")

    def revisarEdad(self):
        if self.__edad >= 80:
            self._morir()
    
    def incrementarEdad(self):
        self.__edad += 1

p1 = Persona('Luis', 10)
print(p1.getNombre())
# p1.__nombre = 20
p1.setNombre('John Persona')
print(p1.getNombre())
print(p1.getEdad())
p1.setEdad(25)
print(p1.getEdad())
p1.revisarEdad()
p1.setEdad(75)
p1.incrementarEdad()
p1.incrementarEdad()
p1.incrementarEdad()
p1.incrementarEdad()
p1.incrementarEdad()
p1.revisarEdad()
```

```python
class EstudianteUP:
    asistencias_totales = 50
    def __init__(self, nombre:str, idx:int) -> None:
        self._nombre = nombre
        self._idx = idx
        self._calificaciones = []
        self._asistencias = 0

    # Setters/getters 
    def getNombre(self):
        return self._nombre
    def setNombre(self, nombre):
        self._nombre = nombre
    def getId(self):
        return self._idx
    def setId(self, idx):
        self.idx = idx
    def getPromedio(self):
        return sum(self._calificaciones) / len(self._calificaciones)
    def addCalificaciones(self, calif):
        if 0 <= calif <= 10:
            self._calificaciones.append(calif)
        else:
            print("Calificación incorrecta")

    def conocerAsistencias(self):
        return self._asistencias
    
    def registrarAsistencia(self):
        self._asistencias += 1
        print("Asistencia registrada!")

    def isAprobado(self):
        promedio = self.getPromedio()
        res = self._asistencias >= self.asistencias_totales * .8 and promedio >= 7.
        if res:
            print(f'El alumno {self._nombre} aprobó con promedio de: {promedio}')
        else:
            print(f'El alumno {self._nombre} reprobó con promedio de: {promedio}')
    # Conocer promedio
    # Conocer asistencias
    # Agregar calificaciones (0, 10)
    # Está aprobado (80% asistencia y el promedio >= 7.0)(50 asistencias total)
    # Registrar asistencia
```
