## Setters/Getters

```python
class Persona:
    def __init__(self, nombre:str, edad:int) -> None:
        self.__nombre = nombre
        self.__edad = edad
    # Este es getter, sirve para devolver el valor
    def getNombre(self):
        return self.__nombre
    # Este es setter, sirve para modificar el valor
    def setNombre(self, nombre:str):
        self.__nombre = nombre

    # Este es getter, sirve para devolver el valor
    def getEdad(self):
        return self.__edad
    # Este es setter, sirve para modificar el valor
    def setEdad(self, edad:int):
        self.__edad = edad

    def _morir(self):
        print(f"QEPD {self.__nombre}!")

    def revisarEdad(self):
        if self.__edad >= 80:
            self._morir()
    
    def incrementarEdad(self):
        self.__edad += 1

p1 = Persona('Luis', 10)
print(p1.getNombre())
# p1.__nombre = 20
p1.setNombre('John Persona')
print(p1.getNombre())
print(p1.getEdad())
p1.setEdad(25)
print(p1.getEdad())
p1.revisarEdad()
p1.setEdad(75)
p1.incrementarEdad()
p1.incrementarEdad()
p1.incrementarEdad()
p1.incrementarEdad()
p1.incrementarEdad()
p1.revisarEdad()
```
