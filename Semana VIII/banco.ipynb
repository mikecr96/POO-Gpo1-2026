```python
"""
atributos de clase: fabricante = "UP"
atributos de instancia: pin, saldo
métodos:
    obtener_pin
    cambiar_pin
    menu 
        cambiar pin (4 dígitos)
        depositar
        transferir
        retirar
        consultar saldo
        salir
"""
class ATM:
    fabricante = "UP"

    def __init__(self, pin:int, saldo:float, nombre:str) -> None:
        self.pin = pin
        self.saldo = saldo
        self.nombre = nombre
        # self.menu()
    def menu(self):
        pin = int(input("Ingrese su PIN, por favor: "))
        if self.obtener_pin(pin):
            print("Entrando a tu cuenta")
        else:
            print("PIN Incorrecto!")
            exit()
        opc = int(input("""Por favor, selecciona una opción.
1. Depositar
2. Transferir
3. Conultar saldo
4. Cambiar PIN
5. Retirar                        
6. Salir                                                
"""))
        if opc == 1:
            self.depositar()
        elif opc == 2:
            self.transferir()
        elif opc == 3:
            self.consultar_saldo()
        elif opc == 4:
            self.cambiar_pin()
        elif opc == 5:
            self.retirar()
        elif opc == 6:
            self.salir()
    def cambiar_pin(self):
        new_pin = int(input("Ingrese su nuevo PIN: "))
        if len(str(new_pin)) == 4 and self.pin != new_pin and isinstance(new_pin, int):
            self.pin = new_pin
            print('PIN válido. Su PIN ha cambiado!')
        else:
            print("PIN Inválido")

    def obtener_pin(self, pin) -> bool:
        return pin == self.pin
    def depositar(self):
        deposit = input("Por favor ingrese la cantidad que quiere depositar: ")
        bandera = False
        try:
            deposit = int(deposit)
            bandera = True
        except ValueError:
            bandera = False
            print("Ingrese solo números")
        finally:
            if bandera: self.saldo += deposit
            else: self.depositar()
            
    def transferir(self):
        other = ATM(1234, 123.50, 'Miguel')
        monto = int(input("Ingrese el monto a transferir"))
        if monto <= self.saldo:
            self.saldo -= monto
            print(f"Enviaste {monto} a {other.nombre}")
            other.saldo += monto
        else:
            print("Saldo insuficiente")
    def retirar(self):
        monto = float(input("Ingrese el monto a retirar: "))
        if monto > self.saldo:
            print("Sado insuficiente")
        else:
            self.saldo -= monto
            print(f'Se ha retirado ${monto} de tu cuenta. Tu nuevo saldo es: ${self.saldo}')
    def consultar_saldo(self):
        print(f'Tu saldo actual es ${self.saldo}.')
    def salir(self):
        exit()
```
