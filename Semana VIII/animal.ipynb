```python
import random
```

```python
class Animal:
    # nombre, genero (M o H), tipo (mamifero, depredador), energia, fuerza
    # isAlive()
    # comer (str) + energia (10)
    # reproducirse -> Animal
    def __init__(self, nombre:str, tipo:str, energia:int, fuerza:int, genero:str) -> None:
        self.nombre = nombre
        self.tipo = tipo
        self.energia = energia
        self.fuerza = fuerza
        self.genero = genero

    def esta_vivo(self):
        return self.energia > 0
    
    def comer(self, comida:str):
        print(f'El animal {self.nombre} comió {comida} y recuperó 10 pts. de energía. Energía actual: {self.energia} pts.')
        self.energia += 10

    def reproducirse(self, other:Animal) -> Animal | None:
        if (self.esta_vivo() and other.esta_vivo()) and (self.genero != other.genero) and (self.tipo == other.tipo):
            print(f'Se creó un nuevo animal de {self.nombre} y {other.nombre}')
            return Animal(f'bebé de {self.nombre} y {other.nombre}', other.tipo, 10, self.fuerza // 2, random.choice(['M', 'H']))
        else:
            print('No fue posible que estos dos animales se reprodujeran.')
            return None
        
    # Método para lógica interna (este no se manda a llamar manualmente)
    def recibir_dano(self, dano):

        self.energia -= dano

        if self.energia < 0:
            self.energia = 0

        print(f'{self.nombre} recibió {dano} de daño, energía actual es: {self.energia}')

    def atacar(self, objetivo:Animal):
        # 33% de probabilidad
        es_critico = random.choice([True, False, False])
        # 30% daño extra si critico
        if es_critico:
            dano_actual = self.fuerza * 1.3
            print("CRÍTICO!!!!!!!!!!!!!")
        else:
            dano_actual = self.fuerza

        print(f'{self.nombre} atacó a {objetivo.nombre}!')
        objetivo.recibir_dano(dano_actual)
```

```python
animales = [
    Animal('Jibanyan', 'mamífero', 20, 20, 'M'),
    Animal('Jeff', 'reptil', 50, 10, 'H'),
    Animal('Pancracio', 'reptil', 200, 50, 'M'),
    Animal('Filemón', 'pez', 1, 2, 'H'),
    Animal('Roberta', 'mamífero', 50, 40, 'H'),
    Animal('Daisy', 'mamífero', 30, 60, 'H')
]

for animal in animales:
    print(animal.nombre)
```
