```python
class Gato:
    # Método especial
    # En otros lenguajes: Método constructor. Método iniciador
    def __init__(self, nombre, color, raza, genero, no_patas=4):
        # Atributos de instancia
        self.nombre = nombre
        self.color = color
        self.raza = raza
        self.genero = genero
        self.no_patas = no_patas

    # Override
    def __str__(self):
        return f'El gato {self.nombre} es de color {self.color}, raza {self.raza}, y es {self.genero}. Tiene {self.no_patas} patas'

    # Método
    def maullar(self): # self SIEMPRE como primer argumento
        print("Meow")

gato1 = Gato("Garfield", 'rojo', 'gordo', 'macho')
gato2 = Gato('Tomas', 'gris', 'medio-tonto', 'macho')
gato3 = Gato('Lucas', 'naranjoso', 'naranja', 'macho')
print(gato1)
print(gato2)
str(gato3)
```

```python
class Motor:
    def __init__(self, cilindros, caballos, torque):
        self.cilindros = cilindros
        self.caballos = caballos
        self.torque = torque
    
    # Override
    def __len__(self):
        return self.cilindros
    
    def __gt__(self, other):
        torque_mayor = self.torque > other.torque
        caballo_mayor = self.caballos > other.caballos
        cilindros_mayor = self.cilindros > other.cilindros
        # Compara si el torque A es mayor que el torque B (other)
        return all([torque_mayor, caballo_mayor, cilindros_mayor])
    -
    def devolverTorque(self):
        return f"Este motor tiene {self.torque} N/m de torque"

motor1 = Motor(4, 30, 210) 
motor2 = Motor(8, 400, 200)
print(motor2.devolverTorque())
len(motor2)
motor2 > motor1
```
