```python
import random
```

```python
class Personaje:
    def __init__(self, nombre:str, vida:int, ataque:int, defensa:int):
        self.nombre = nombre
        self.vida_max = vida
        self.vida_actual = self.vida_max
        self.ataque = ataque
        self.defensa = defensa

    def isAlive(self):
        return self.vida_actual > 0
    
    def curar(self, cura):
        if cura > self.vida_max - self.vida_actual:
            cura = self.vida_max - self.vida_actual
        if self.vida_actual != self.vida_max and self.isAlive():
            self.vida_actual += cura
        # if self.vida_actual > self.vida_max:
        #     self.vida_actual = self.vida_max        

        print(f'{self.nombre} sanó {cura} pts. de vida. Vida actual es: {self.vida_actual}')

    # Método para lógica interna (este no se manda a llamar manualmente)
    def recibir_dano(self, dano):
        dano_real = dano - self.defensa
        if dano_real <= 0:
            dano_real = 3

        self.vida_actual -= dano_real

        if self.vida_actual < 0:
            self.vida_actual = 0

        print(f'{self.nombre} recibió {dano_real} de daño, vida actual es: {self.vida_actual}')

    def atacar(self, objetivo:Personaje):
        # 33% de probabilidad
        es_critico = random.choice([True, False, False])
        # 30% daño extra si critico
        if es_critico:
            dano_actual = self.ataque * 1.3
            print("CRÍTICO!!!!!!!!!!!!!")
        else:
            dano_actual = self.ataque

        print(f'{self.nombre} atacó a {objetivo.nombre}!')
        objetivo.recibir_dano(dano_actual)
        
    
    def __str__(self) -> str:
        return f'''Nombre: {self.nombre}
Vida actual: {self.vida_actual}
Vida máxima: {self.vida_max}
Ataque: {self.ataque}
Defensa: {self.defensa}
'''
```

```python
# Simulación del juego

P1 = Personaje('Héroe', 200, 80, 100)
P2 = Personaje('Duende 1', 50, 50, 10)
P3 = Personaje('Duende 2', 2000, 1, 2)
print(P1, P2, P3)
```
