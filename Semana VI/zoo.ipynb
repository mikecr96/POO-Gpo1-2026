```python
# NOTA: No hemos visto nada de lo que viene a continuación, por favor, no se espanten!

# Clase, atributos, métodos, objetos, propiedades, polimorfismo, etc...
# Clases son templates

class Animal:
    # Método 1 (los métodos son funciones, pero se llaman diferente cuando están dentro de una clase)
    def __init__(self, nombre, edad, especie, sonido):
        self.nombre = nombre
        self.edad = edad
        self.especie = especie
        self.sonido = sonido

    # Método 2
    def hacer_sonido(self):
        print(f"{self.nombre} hace un {self.sonido}")


leon = Animal("Simba", 10, "León", "Rugido") # objeto tipo Animal
elefante = Animal("Elefantito", 1, "Elefante", "Trompeta") # objeto tipo Animal
mono = Animal("Morph", 2, "mono", "Aúllido") # objeto tipo Animal

# dot-notation

leon.hacer_sonido()
elefante.hacer_sonido()
mono.hacer_sonido()
```
