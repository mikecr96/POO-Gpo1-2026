## List Comprehension

```python
# Hacer una lista con los números del 1 al 100
numeros = []

for i in range(1, 101):
    numeros.append(i)

print(numeros)
```

```python
numeros = [i for i in range(1, 101)]
print(numeros)
```

## Dict comprehension

```python
# Diccionario [1, 100] donde la llave es el número, pero en string y el valor es el núm (int)
diccionario = {}

for i in range(1, 101):
    diccionario[f'{i}'] = i

diccionario
```

```python
# Diccionario [1, 100] donde la llave es el número, pero en string y el valor es el núm (int)
diccionario = {str(i):i for i in range(1, 101)}
print(diccionario)
```

## Matrices (simulación)

```python
mat3x3 = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

type(mat3x3)
```

```python
len(mat3x3)
```

```python
# fi-co

mat3x3[1][1]
```

```python
mat3x3[2][2]
```

```python
for i in mat3x3:
    print(i)
```

```python
for f in mat3x3:
    for c in f:
        print(c)
```

```python
for f in range(3):
    for c in range(3):
        # 1) f = 0 y c = 0
        # 2) f= 0 y c = 1
        # 3) f= 0 y c = 2
        # 4) f= 1 y c = 0
        # 5) f= 1 y c = 1
        # 6) f= 1 y c = 2
        print(mat3x3[f][c])
```

```python
from random import randint

# Matriz de 10x5

mat10x5 = []
for f in range(10):
    mat10x5.append([randint(1, 100) for _ in range(5)])

print(mat10x5)
```

```python
for f in mat10x5:
    for c in f:
        print(f'{c}', end=' ')
    print()
```

```python
for f in range(len(mat10x5)):
    for c in range(len(mat10x5[0])):
        print(f'{mat10x5[f][c]}', end=' ')
    print()
```
