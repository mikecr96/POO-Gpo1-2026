```python
# cta1.nip -> nip
# cta1.nip = 1234
class CuentaBancaria:
    def __init__(self, cliente:str, nip:int) -> None:
        self._cliente = cliente
        self._nip = nip
        self._saldo = 0 # Name mangling

    def getCliente(self):
        return self._cliente
    
    def setCliente(self, cliente):
        self._cliente = cliente

    # getterActualizado
    @property # Decoradores
    def nip(self):
        return self._nip
    
    #setterActualizado 
    @nip.setter
    def nip(self, nip):
        self._nip = nip

cta1 = CuentaBancaria('Rodrigo', 7890)
print(cta1.nip)
cta1.nip = 1234
print(cta1.nip)
```

```python
class Mochila:
    def __init__(self, peso_max=25) -> None:
        self.peso_max = peso_max
        self.peso_actual = 0
        self.contenido = []

    def meterArticulo(self, articulo:Articulo):
        if articulo.peso + self.peso_actual > self.peso_max:
            print(f"{articulo.nombre} no cabe en mi mochilita")
        else:
            self.peso_actual += articulo.peso
            self.contenido.append(articulo)
            print(f"{articulo.nombre} entró en mi mochilita")

    def quitarArticulos(self, articulo:Articulo):
        if articulo in self.contenido:
            self.contenido.remove(articulo)
            self.peso_actual -= articulo.peso
            print(f'El artículo {articulo.nombre} fue removido de la mochila.')
        else:
            print("Ese artículo no está en la mochila.")

    def __str__(self) -> str:
        if len(self.contenido) == 0:
            return 'Esta mochila está vacía'
        else:
            string = f'Esta mochila tiene {len(self.contenido)} artículos dentro. Peso ocupado: {self.peso_actual:.3f} kg.\n'
            # for i in self.contenido:
            #     string += f'Artículo: {i._nombre}. Peso {i._peso} kg.\n'
            string += '\n'.join([f'{n}.- Artículo: {i._nombre}. Peso {i._peso} kg.' for n, i in enumerate(self.contenido, 1)])
            string += f'\nCapacidad disponible: {(self.peso_max - self.peso_actual):.3f} kg.'
            return string

class Articulo:
    def __init__(self, nombre:str, peso:float) -> None:
        self._nombre = nombre
        self._peso = peso
    @property
    def nombre(self):
        return self._nombre
    @property
    def peso(self):
        return self._peso
    def __str__(self) -> str:
        return f'El artículo {self._nombre} pesa {self._peso} kg.'

mochila = Mochila()
lapiz = Articulo('lápiz', .2)
mac = Articulo('MacBook', 3.1)
mochila.meterArticulo(lapiz)
```
