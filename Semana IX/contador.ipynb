SETTERS/GETTERS

Actualizado

```python
# cta1.nip -> nip
# cta1.nip = 1234
class CuentaBancaria:
    def __init__(self, cliente:str, nip:int) -> None:
        self._cliente = cliente
        self._nip = nip
        self._saldo = 0 # Name mangling

    def getCliente(self):
        return self._cliente
    
    def setCliente(self, cliente):
        self._cliente = cliente

    # getterActualizado
    @property # Decoradores
    def nip(self):
        return self._nip
    
    #setterActualizado 
    @nip.setter
    def nip(self, nip):
        self._nip = nip

cta1 = CuentaBancaria('Rodrigo', 7890)
print(cta1.nip)
cta1.nip = 1234
print(cta1.nip)
```
