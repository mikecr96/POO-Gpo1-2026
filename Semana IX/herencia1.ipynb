```python
class Animal: 
    def __init__(self, edad, especie) -> None:
        self.edad = edad
        self.especie = especie

    def hablar(self):
        print('Este animal habló')
class Perro(Animal): pass # Perro hereda (o extiende) de Animal
class Vaca(Animal): pass
class Gato(Animal): 
    # Override
    def __init__(self, color) -> None:
        self.color = color

g2 = Gato(2, 'gato naranja')
g2.hablar()
```
