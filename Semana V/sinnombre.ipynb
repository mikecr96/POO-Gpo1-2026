## Método 2

La ventaja de este método es que no tienes que dedicar una línea a cerrar

```python
with open("/Users/miguelcamargorojas/Documents/UP/POO-Gpo1-2026/Semana V/test.txt", 'r') as t:
    print(t.read())
```

```python
with open("/Users/miguelcamargorojas/Documents/UP/POO-Gpo1-2026/Semana V/test.txt", 'w') as t:
    t.write(texto)
```

```python
with open("/Users/miguelcamargorojas/Documents/UP/POO-Gpo1-2026/Semana V/test.txt", 'a') as t:
    t.write('Añadiendo saluditos.\nXD')
```

```python
textos = ["Hola mundo\n", "Saludos\n", "Alumnos de POO\n", "Los tqm\n", "Arriba la UP"]

with open("/Users/miguelcamargorojas/Documents/UP/POO-Gpo1-2026/Semana V/test.txt", 'w') as t:
    t.writelines(textos)
```

```python
with open("/Users/miguelcamargorojas/Documents/UP/POO-Gpo1-2026/Semana V/test.txt", 'r') as t:
    lista_texto = t.readlines()

lista_texto
```
