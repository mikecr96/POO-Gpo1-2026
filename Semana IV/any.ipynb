```python
lista_alumnos = ['Miguel', 'Andrea', 'Fernanda', 'Juan', 'Ximena', 'Pedrito']
"""
1 Miguel
2 Andrea
3 Fernanda...
"""

for i in range(len(lista_alumnos)):
    print(f'{i+1} {lista_alumnos[i]}')
```
