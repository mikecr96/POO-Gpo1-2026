# Emulando matrices

```python
# Matriz de 2x2
m1 = [
    [1, 2], 
    [3, 4]
    ]

print(type(m1))
```

```python
# Esto no se puede hacer 
# m1[0, 0]
```

```python
# mnemónico
# fi-co : fila, columna
m1[0][0]
```

```python
m1[1][1]
```

```python
# Matriz de 3x3
m2 = [
    [1, 2, 3], 
    [4, 5, 6],
    [7, 8, 9]
]

m2[1][1]
```

```python
m2[2][2]
```

```python
m2[0][2]
```

```python
m2
```

```python
for i in m2:
    print(i)
```

```python
for i in range(3):
    print(m2[i][i])
```

# For anidado (nested for)

```python
for f in m2: # For más externo
    for c in f: # For más interno
        print(f, c)
```

```python
for f in m2: # For más externo
    for c in f: # For más interno
        print(c)
```

```python
for f in range(3):
    for c in range(3):
        print(m2[f][c], end="...")
```

```python
for f in range(3):
    for c in range(3):
        print(f"Fila: {f} , columna: {c}", end=" ")
    print()
```

```python
for f in range(3):
    for c in range(3):
        print(f"({f},{c})", end=" ")
    print()
```
