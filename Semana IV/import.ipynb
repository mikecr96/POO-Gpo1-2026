# Emulando matrices

```python
# Matriz de 2x2
m1 = [
    [1, 2], 
    [3, 4]
    ]

print(type(m1))
```

```python
# Esto no se puede hacer 
# m1[0, 0]
```

```python
# mnemónico
# fi-co : fila, columna
m1[0][0]
```

```python
m1[1][1]
```

```python
# Matriz de 3x3
m2 = [
    [1, 2, 3], 
    [4, 5, 6],
    [7, 8, 9]
]

m2[1][1]
```

```python
m2[2][2]
```

```python
m2[0][2]
```

```python
m2
```

```python
for i in m2:
    print(i)
```

```python
for i in range(3):
    print(m2[i][i])
```

# For anidado (nested for)

```python
for f in m2: # For más externo
    for c in f: # For más interno
        print(f, c)
```

```python
for f in m2: # For más externo
    for c in f: # For más interno
        print(c)
```

```python
for f in range(3):
    for c in range(3):
        print(m2[f][c], end="...")
```

```python
for f in range(3):
    for c in range(3):
        print(f"Fila: {f} , columna: {c}", end=" ")
    print()
```

```python
for f in range(3):
    for c in range(3):
        print(f"({f},{c})", end=" ")
    print()
```

```python
for f in m2: # For más externo
    for c in f: # For más interno
        print(f"{c * 10}", end=" ")
    print()
```

# random
Nota: no existe lo aleatorio, en realidad es pseudo-aleatorio

```python
import random
```

```python
random.random()
```

```python
random.randint(0, 10)
```

```python
lista_alumnos = ['Randal', 'Emilio', 'Antonio', 'Juanpa', 'Axel', 'Adriana']

random.choice(lista_alumnos)
```

```python
random.shuffle(lista_alumnos)
```

```python
lista_alumnos
```

```python
random.sample(lista_alumnos, 3)
```

```python
lista = []

# Llenar la lista con 100 valores enteros aleatorios
for i in range(101):
    lista.append(random.randint(0, 100))

print(lista)
```

```python
# Nota, cuando no se ocupe la variable iteradora, ocupamos '_'
lista = [random.randint(0, 100) for _ in range(101)]

print(lista)
```

```python
# Matriz de 5x5 con enteros aleatorios [1, 100]
m5x5 = []

for f in range(5):
    m5x5.append([random.randint(0, 100) for _ in range(5)])    

m5x5
```

```python
# Matriz de 5x5 con enteros aleatorios [1, 100]
m5x5 = [[random.randint(0, 100) for _ in range(5)] for _ in range(5)]

# for f in range(5):
#     m5x5.append([random.randint(0, 100) for _ in range(5)])    

m5x5
```

```python
# Crear una función que simule lanzar un dado: [1, 6]

def tirarDado():
    print(f'Tiraste el dado y cayó: {random.randint(1, 6)}')
```

```python
tirarDado()
```

```python
# Crear una función que simule lanzar una moneda [cara, cruz] random.random < .5 cara >.5 cruz
def lanzarMoneda():
    res = random.random()
    if res >= .5:
        print('Cayó cruz')
    else:
        print('Cayó cara')

lanzarMoneda()
```
