```python
conjunto = {1, 1, 1, 2, 2, 3, 3, 4}
conjunto
```

```python
len(conjunto)
```

```python
for pepito in conjunto:
    print(pepito)
```

```python
dir(conjunto)
```

```python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}
```

```python
# a union b
# a.union(b)

a.union(b)
```

```python
a | b # | oa para union entre dos conjuntos
```

```python
# a intersección b
# a.intersection(b)

a.intersection(b)
```

```python
a & b # & oa para intersection entre dos conjuntos
```

```python
a.symmetric_difference(b)
```

```python
a.difference(b)
```

```python
a - b # oa - es el atajo diferencia
```
