```python
"""
try:
    {bloque}
<except>:
    {bloque}
<finally>:
    {bloque}
"""
try:
    10 / 0
except:
    print("Error!")
finally: # finally SIEMPRE se ejecuta
    print("Hola mundo")
```

```python
try:
    print(11 / 2)
except:
    print("Error!")
finally: # finally SIEMPRE se ejecuta
    print("Hola mundo")
```

```python
try:
    print(11 / 0)
# except ZeroDivisionError:
#     print("Error, no se permite la división entre 0")
except TypeError:
    print("Error de tipo")
except ValueError:
    pass # No hace nada
finally: # finally SIEMPRE se ejecuta
    print("Hola mundo")
```
