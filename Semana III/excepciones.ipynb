```python
"""
try:
    {bloque}
<except>:
    {bloque}
<finally>:
    {bloque}
"""
try:
    10 / 0
except:
    print("Error!")
finally: # finally SIEMPRE se ejecuta
    print("Hola mundo")
```

```python
diccionario = {1:'hola'}

try:
    # pass
    # print(11 / 0)
    diccionario['mundo']
except ZeroDivisionError as e:
    print(f"Error, no se permite la división entre 0 | {e}")
except TypeError as e:
    print("Error de tipo")
except ValueError as e:
    pass # No hace nada
except KeyError as e:
    print(f"Esta llave no existe: {e}")
else: # Solo se entra al else cuando no se cacha la excepción
    print("Hola desde el else")
finally: # finally SIEMPRE se ejecuta
    print("Hola mundo")
```
