# Comprehension

```python
# list comprehension
# Llenar la lista con elementos del 1 al 100

lista = []

for i in range(1, 101):
    lista.append(i) # append es el método que mete un valor al final de la lista

print(lista)
```

```python
# list comprehension
# Llenar la lista con elementos del 1 al 100

lista2 = [x for x in range(1, 101)]

# for i in range(1, 101):
#     lista.append(i) # append es el método que mete un valor al final de la lista

print(lista2)
```

```python
# set comprehension
# Llenar la lista con elementos del 1 al 100

conjunto = {x for x in range(1, 101)}

# for i in range(1, 101):
#     lista.append(i) # append es el método que mete un valor al final de la lista

print(conjunto)
```

```python
# tupla comprehension
# Llenar la lista con elementos del 1 al 100

tupla = (x for x in range(1, 101)) # Esto NO genera una tupla, genera un generador

# for i in range(1, 101):
#     lista.append(i) # append es el método que mete un valor al final de la lista

tupla
```

```python
type(tupla)
```

```python
for i in tupla: print(i)
```

```python
# dict comprehension
diccionario = {x: x for x in range(1, 101)}

diccionario
```

```python
# dict comprehension
# {
#     palabra : len(palabra)
# }

diccionario = {i: len(i) for i in ["hello", "world"]}

diccionario
```

```python
[s for s in "hola mundo"]
```

```python
#
lista = [i for i in range(1, 101)] # Quitar pares

# Preguntar cuáles son pares
for i in lista:
    if i % 2 == 0:
        print(i)
```

```python
# Llenar con valores del 1 al 100, solo si son impares
# Itera en el rango 1 a 100, y agrega a la lista solo aquellos elementos 
# que cumplan la condición
lista = [i for i in range(1, 101) if i % 2 != 0]
lista
```
